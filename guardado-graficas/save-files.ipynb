```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df_ventas = pd.read_csv('../data/tienda_procesado.csv', index_col='fecha')
df_ventas
```

 **`Guardando imagenes con Matplotlib`**

`.figure.savefig()`

```python
plt.style.use('ggplot')
figure_matplotlib = df_ventas.plot.line(
    figsize=(10, 5), # Tamao de la grfica (ancho, alto)
    title = 'Ventas de tres tipos de productos en 2015' # aqu definimos el ttulo general de la grfica
)

# para poner ttulo a los ejes
# eje x

figure_matplotlib.set_xlabel(
    'Perodo',

# por defecto aparece en el centro pero puedo cambiar su localizacin, no es recomendable pero si posible hacerlo
    loc='left'
)

# eje y
figure_matplotlib.set_ylabel(
    'Ventas (en millones de dolares)',
    )

```

```python
# AQU OCURRE EL DESPEGUEEE
figure_matplotlib.figure.savefig('../files/ventas_2015.png', format= 'png', dpi=800) #con dpi aumentamos la calidad del archivo
```

 **`Aadiento imagenes con seaborn`**

`.figure.savefig()`

```python
plt.figure(figsize=(8, 4)
)
sns.set_style('ticks')
sns.set_theme('paper')

figure_seaborn = sns.lineplot(
    data=df_ventas,
)

# para poner ttulo general !!atencin esto a afecta a nivel general a cualquier grfica que pueda generar yo depus con seaborn
plt.title('Ventas de tres tipos de productos en 2015')
plt.xlabel('Perodo')
plt.ylabel('Ventas (en millones de dolares)')

# una manera ms localizada y especifica
figure_seaborn.set(
    title='Ventas de tres tipos de productos en 2015 (modo especfico)',
    xlabel='Perodo (modo especfico)',
    ylabel='Ventas (en millones de dolares) (modo especfico)'
)
```

```python
# AQU OCURRE EL DESPEGUEEE
figure_seaborn.figure.savefig('../files/ventas.png', format='png', dpi=300, bbox_inches='tight') #con bbox_inches aade todo lo que es visible, cuando aadimos elementos extra por fuera del canvas tendremos que utilizarlo
```

 **`Guardando grficas con plotly`**

En la esquina superior derecha pulsamos sobre el icono de la cmara y de descarga automaticamente el archivo, esto en el caso de plotly tiene inconvenientes porqu perdomos toda la interactividad

```python
figure_plotly = px.line(
    df_ventas,
    width= 600,
    height = 300,
    template = 'plotly_dark',
    # para poner ttulo a la grfica
    title = 'Ventas de tres tipos de productos en 2015',
    # para poner ttulo a los ejes de la grfica
    labels = {
                'fecha': 'Perodo', 
                'value': 'Ventas (en millones de dolares)',
                # para modificar la leyenda
                'variable': 'Tipo de producto'
            }
)

# tambin se pueden modificar los ttulos de los ejes de manera ms intuitiva llamando a los mtodos especficos
figure_plotly.update_layout(
    xaxis_title='Perodo (especfico)',
    yaxis_title='Ventas (en millones de dolares) (especfico)'
)

figure_plotly.show()
```
