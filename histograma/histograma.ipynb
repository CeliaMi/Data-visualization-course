```python
import pandas as pd
titanic_df = pd.read_csv('../grafica-barras/titanic.csv')
```

```python
# para que elimine los valores Nulos de la columna Age, cuando yo le digo axis= ' index' le estoy diciendo que elimine la fila y no la columna
titanic_df = titanic_df.dropna(subset=[ 'Age'], axis = 'index')
titanic_df
```

```python
titanic_df.hist(
    'Age',
)
```

```python
titanic_df.hist(
    'Age',
    bins = 20
)
```

```python
titanic_df.hist(
    'Age',
    bins = 4
)

# Si ponemos pocos bins perdemos muchos detalles de los valores y podemos sacar conclusiones erroneas y disfrazar la distribucin
```

```python
titanic_df.hist(
    'Age',
    by = 'Sex',
    bins = 20
)
```

# GRFICO DE KDE ( KERNEL DENSITY ESTIMATOR)
Es una curva suavizada que nos permite entender la forma de nuestra distribucin


```python
import seaborn as sns
```

```python
sns.histplot(
    titanic_df,
    x= 'Age',
    kde = True
)
```

```python
sns.histplot(
    titanic_df,
    x= 'Age',
    bins= 4,
    kde = True
)

# el KDE no se ve afectado por los bins
# poner la linea de KDE tambin nos ayuda a saber el nmero optimo de bins
```

```python
sns.histplot(
    titanic_df,
    x= 'Age',
    kde = True,
    stat = 'probability'
)
```

```python
sns.histplot(
    titanic_df,
    x= 'Age',
    kde = True,
    bins= 20,
    hue= 'Sex'
)
```

```python
sns.histplot(
    titanic_df,
    x= 'Age',
    cumulative = True,
    bins= 20,
    fill=False,
)
```

```python
sns.histplot(
    titanic_df,
    x= 'Age',
    cumulative = True,
    bins= 20,
    fill=False,
    element= 'step'
)
```
