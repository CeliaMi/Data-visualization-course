```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df_ventas = pd.read_csv('../data/tienda_procesado.csv', index_col='fecha')
df_ventas
```

 **`AÃ±adiento estilo al texto con Matplotlib`**

```python
plt.style.use('ggplot')
figure_matplotlib = df_ventas.plot.line(
    figsize=(10, 5), # TamaÃ±o de la grÃ¡fica (ancho, alto)
    title = 'Ventas de tres tipos de productos en 2015' # ðaquÃ­ definimos el tÃ­tulo general de la grÃ¡fica
)

# para poner tÃ­tulo a los ejes
# eje xð

figure_matplotlib.set_xlabel(
    'PerÃ­odo',

# por defecto aparece en el centro pero puedo cambiar su localizaciÃ³n, no es recomendable pero si posible hacerlo
    loc='left'
)

# eje yð
figure_matplotlib.set_ylabel(
    'Ventas (en millones de dolares)',
        fontdict = {        # ððaquÃ­ definimos el estilo de la fuente
            'name': 'Arial',
            'color': 'red',
            'size': 15
        }
    )

```

Si yo quiero seÃ±alizar un punto concreto y poner informaciÃ³n adicional, utilizo el mÃ©todo annotate

```python
from IPython.display import display


figure_matplotlib.annotate(
    'Noviembre OJO!',
    xy=('2015-11-01', 700000),
)

display(figure_matplotlib)
plt.show()

```

 **`AÃ±adiento tÃ­tulos con seaborn`**

```python
plt.figure(figsize=(8, 4)
)
sns.set_style('ticks')
sns.set_theme('paper')

figure_seaborn = sns.lineplot(
    data=df_ventas,
)

# para poner tÃ­tulo general !!atenciÃ³n esto a afecta a nivel general a cualquier grÃ¡fica que pueda generar yo depuÃ©s con seaborn
plt.title('Ventas de tres tipos de productos en 2015')
plt.xlabel('PerÃ­odo')
plt.ylabel('Ventas (en millones de dolares)')

# una manera mÃ¡s localizada y especifica
figure_seaborn.set(
    title='Ventas de tres tipos de productos en 2015 (modo especÃ­fico)',
    xlabel='PerÃ­odo (modo especÃ­fico)',
    ylabel='Ventas (en millones de dolares) (modo especÃ­fico)'
)
```

 **`AÃ±adiento tÃ­tulos con plotly`**

```python
figure_plotly = px.line(
    df_ventas,
    width= 600,
    height = 300,
    template = 'plotly_dark',
    # para poner tÃ­tulo a la grÃ¡ficað
    title = 'Ventas de tres tipos de productos en 2015',
    # para poner tÃ­tulo a los ejes de la grÃ¡ficað
    labels = {
                'fecha': 'PerÃ­odo', 
                'value': 'Ventas (en millones de dolares)',
                # para modificar la leyenda
                'variable': 'Tipo de producto'
            }
)

# tambiÃ©n se pueden modificar los tÃ­tulos de los ejes de manera mÃ¡s intuitiva llamando a los mÃ©todos especÃ­ficos
figure_plotly.update_layout(
    xaxis_title='PerÃ­odo (especÃ­fico)',
    yaxis_title='Ventas (en millones de dolares) (especÃ­fico)'
)

figure_plotly.show()
```
