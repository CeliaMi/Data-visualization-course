## Pandas
**`tamaño de la gráfica`**

```python
plt.style.use('Solarize_Light2')
df_ventas.plot.line(
    figsize=(10, 5) # Tamaño de la gráfica (ancho, alto)
)
```

**`color de la cuadrícula`**

Para poder editar la cuadrícula (`grid`) necesitamos guardar la gráfica en una variable para que python recoja el objeto generado y así poder llamarlo y hacer modificaciones sobre ella.
En este caso hemos llamado a nuestra gráfica con el nombre de `figure`

```python
plt.style.use('Solarize_Light2')
# Ponemos la gráfica en una variable para poder modificarla
figure = df_ventas.plot.line(
    figsize=(10, 5) 
)
# Modificamos el color de la línea llamando a mi variable
#cambiamos el color del eje x
figure.xaxis.grid(
    color='blue'
)

#eje y 
figure.yaxis.grid(
    color='red'
)
```

**`estilo de la línea de la cuadrícula`**


[matplotlib line_style📖](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)
<img src='../files/sphx_glr_linestyles_001_2_00x.webp' alt='schema lines styles on matplotlib'>


Añadimos el atributo `linestyle` a nuestro objeto, especificando el eje 

```python
plt.style.use('Solarize_Light2')
figure_line_style = df_ventas.plot.line(
    figsize=(10, 5) 
)
# Modificamos el estilo de la línea llamando a mi variable a través del atributo linestyle
figure_line_style.grid(
    linestyle='dashed',
     # coloreamos las líneas de la cuadrícula de negro para verlas mejor
     color='black'
)
```
