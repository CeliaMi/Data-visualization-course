```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df_ventas = pd.read_csv('../data/tienda_procesado.csv', index_col='fecha')
df_ventas
```

A lo largo de este notebook vamos a ver las diferentes maneras de ajustar el tamaño, el color y el estilo de línea de los gráficos dentro de **Matplotlib, pandas, seaborn y plotly**

**las gráficas de Pandas y de Seaborn estan construidas sobre matplotlib**, es decir cuando utilizamos estas librerías por detrás de base esta matplotlib, sin embargo plotly es otra librería completamente independiente.

## Pandas
**`tamaño de la gráfica`**

```python
plt.style.use('Solarize_Light2')
df_ventas.plot.line(
    figsize=(10, 5) # Tamaño de la gráfica (ancho, alto)
)
```

**`color de la cuadrícula`**

Para poder editar la cuadrícula (`grid`) necesitamos guardar la gráfica en una variable para que python recoja el objeto generado y así poder llamarlo y hacer modificaciones sobre ella.
En este caso hemos llamado a nuestra gráfica con el nombre de `figure`

```python
plt.style.use('Solarize_Light2')
# Ponemos la gráfica en una variable para poder modificarla
figure = df_ventas.plot.line(
    figsize=(10, 5) 
)
# Modificamos el color de la línea llamando a mi variable
#cambiamos el color del eje x
figure.xaxis.grid(
    color='blue'
)

#eje y 
figure.yaxis.grid(
    color='red'
)
```

**`estilo de la línea de la cuadrícula`**


[matplotlib line_style📖](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)

<img src='../files/sphx_glr_linestyles_001_2_00x.webp' alt='schema lines styles on matplotlib'>


Añadimos el atributo `linestyle` a nuestro objeto, especificando el eje 

```python
plt.style.use('Solarize_Light2')
figure_line_style = df_ventas.plot.line(
    figsize=(10, 5) 
)
# Modificamos el estilo de la línea llamando a mi variable a través del atributo linestyle
figure_line_style.grid(
    linestyle='dashed',
     # coloreamos las líneas de la cuadrícula de negro para verlas mejor
     color='black'
)
```

## Seaborn
Podemos hacer lo mismo que en el caso anterior con pandas, guardar la gráfica en un objeto para seguir realizando cambios.
Dentro de Seaborn hay otra manera de hacerlo a través de matprolib que creará métodos globales que se aplicarán en todos mis gráficos

```python
plt.figure(
    figsize=(10,3)
)
sns.set_theme()
sns.set_style("dark")

sns.lineplot(
    data=df_ventas
)
```

## Plotly
Esta libería no se basa en matplotlib

```python
px.line(
    df_ventas,
    width= 600,
    height = 300,
    template = 'plotly_dark'
)
```

`update_xaxes`, `updateyaxes` para seleccionar los ejes a modificar.

Los atributos que podemos modificar son infinitos:
`gridcolor`, `gridwidth`, `griddash`, `minor_griddash`,.... 

```python
figure_plotly = px.line(
    df_ventas,
    width= 600,
    height = 300,
    template = 'plotly_dark'
)

# update_xaxes para modificar el eje x
figure_plotly.update_xaxes(showgrid =False)
```
