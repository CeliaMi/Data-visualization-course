```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df_ventas = pd.read_csv('../data/tienda_procesado.csv', index_col='fecha')
df_ventas
```

 **`Añadiento títulos con Matplotlib`**

```python
plt.style.use('ggplot')
figure = df_ventas.plot.line(
    figsize=(10, 5), # Tamaño de la gráfica (ancho, alto)
    title = 'Ventas de tres tipos de productos en 2015' # 👈aquí definimos el título general de la gráfica
)

# para poner título a los ejes
# eje x👇

figure.set_xlabel(
    'Período',

# por defecto aparece en el centro pero puedo cambiar su localización, no es recomendable pero si posible hacerlo
    loc='left'
)

# eje y👇
figure.set_ylabel(
    'Ventas (en millones de dolares)',
    )

```

 **`Añadiento títulos con seaborn`**

```python
plt.figure(figsize=(8, 4)
)
sns.set_style('ticks')
sns.set_theme('paper')

figure = sns.lineplot(
    data=df_ventas,
)

# para poner título general !!atención esto a afecta a nivel general a cualquier gráfica que pueda generar yo depués con seaborn
plt.title('Ventas de tres tipos de productos en 2015')
plt.xlabel('Período')
plt.ylabel('Ventas (en millones de dolares)')

# una manera más localizada y especifica
figure.set(
    title='Ventas de tres tipos de productos en 2015 (modo específico)',
    xlabel='Período (modo específico)',
    ylabel='Ventas (en millones de dolares) (modo específico)'
)
```

 **`Añadiento títulos con plotly`**

```python
figure_plotly = px.line(
    df_ventas,
    width= 600,
    height = 300,
    template = 'plotly_dark',
    # para poner título a la gráfica👇
    title = 'Ventas de tres tipos de productos en 2015',
    # para poner título a los ejes de la gráfica👇
    labels = {
                'fecha': 'Período', 
                'value': 'Ventas (en millones de dolares)',
                # para modificar la leyenda
                'variable': 'Tipo de producto'
            }
)

# también se pueden modificar los títulos de los ejes de manera más intuitiva llamando a los métodos específicos
figure_plotly.update_layout(
    xaxis_title='Período (específico)',
    yaxis_title='Ventas (en millones de dolares) (específico)'
)

figure_plotly.show()
```
