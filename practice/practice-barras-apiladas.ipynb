### 1️⃣ Cargamos librerías y el archivo cvs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('..\data\datasets_top_4.csv',index_col='dataset')
df
```

Dividimos el dataset en dos:

```python
# iloc divide el dataframe por filas desde la primera hasta la 14 y desde la 14 hasta la última 
df1 = df.iloc[:14,:]
df2 = df.iloc[14:,:]
```

```python
# una figura 2 columnas ( grafico 1 columna 1 gráfico 2 columna 2)
fig, (ax1, ax2) = plt.subplots(1, 2)

fig.set_size_inches(7, 4)
# .barh es para hacer un gráfico de barras horizontal
df1.plot.barh(
    # stacked=True es para que las barras se apilen
    stacked=True,
    ax=ax1,
    width=0.7,
    color={
        'cat1': '#000086',
        'cat2': '#5396f4',
        'cat3': '#a1dbe8',
        'otros': '#DADADA'
    }
)

df2.plot.barh(
    stacked=True,
    ax=ax2,
    width=0.7,
    color={
        'cat1': '#000086',
        'cat2': '#5396f4',
        'cat3': '#a1dbe8',
        'otros': '#DADADA'
    }
)

ax2.set_ylabel('')
ax1.set_ylabel('Dataset')

ax2.set_xlabel('Porcentaje')
ax1.set_xlabel('Porcentaje')

ax1.legend().remove()

ax2.legend(
    ['Categoría 1', 'Categoría 2', 'Categoría 3', 'Otros'],
    ncol=4,
    loc="upper left",
    bbox_to_anchor=(-1.3, -0.15)
)

ax2.yaxis.tick_right()

ax1.text(
    75, -5,
    "*Algunos datasets han sido ignorados del análisis",
    fontsize=6,
    bbox={
        "facecolor": "gray",
        "alpha": 0.2,
        "pad": 5,
    }
)

fig.suptitle('Distribución de las categorias dentro de los datasets', fontsize=12)
```
